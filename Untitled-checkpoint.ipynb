```python
from uuid import *
from math import *
```

```python
def operation(x,y,op:str):
    match(op):
        case "+":
            print(f"{x}+{y}={x+y}")
            
        case "-":
            print(f"{x}-{y}={x-y}")
        case "%":
            print(f"{x}%{y}={x%y}")
        case "÷":
            print(f"{x}÷{y}={x/y}")
        case "×":
            print(f"{x}×{y}={x*y}")
        case "*":
            print(f"{x}*{y}={x*y}")
        case "^":
            print(f"{int(x)}^{int(y)}={int(x)^int(y)}")
        case "/":
            print(f"{int(x)}/{int(y)}={(x)/(y)}")
        case "**":
            print(f"{x}**{int(y)}={x**y}")
        case "//":
            print(f"{int(x)}//{int(y)}={(x)//(y)}")
```

```python
while True:
    try:
        a=eval(input("Give any Number:"))
        b=eval(input("Give any another Number:"))
        op=input("Give any Operator:")
        result=operation(a,b,op)

    except Exception as e:
        print("Error Occured!")
```
